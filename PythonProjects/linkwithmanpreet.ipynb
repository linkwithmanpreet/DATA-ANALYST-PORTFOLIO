# **Car Dataset Analysis**

In this notebook, we will perform various data analysis and visualization tasks on a car dataset. The dataset is stored in a CSV file, and we'll use pandas, seaborn, and matplotlib for our analysis.

## **Step 1: Import Libraries**

We begin by importing the necessary libraries.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## **Step 2: Load the Dataset**

We load the car dataset from the specified file path and take a first look at the data.


```python
# Import dataset
Car_data = pd.read_csv(r'C:\Users\MYSHA\OneDrive\Desktop\linkwithmanpreet\cars_data.csv')

# Display the first few rows of the dataset
Car_data.head()
```

## **Step 3: Explore the Dataset**

Next, we'll explore the dataset by checking its shape, columns, and data types.


```python
# Check the shape of the dataset
Car_data.shape

# List the column names
Car_data.columns

# Get information about the dataset
Car_data.info()

# Check for null values
Car_data.isnull()
```

## **Step 4: Handle Missing Values**

We'll identify the number of missing values in each column and handle them accordingly. In this case, we'll fill missing values in the 'Cylinders' column with the mean value.


```python
# Find the number of null values in each column
Car_data.isnull().sum()

# Assign the result back to the original column without using inplace
Car_data['Cylinders'] = Car_data['Cylinders'].fillna(Car_data['Cylinders'].mean())

# Verify that missing values have been filled
Car_data.isnull().sum()

# Display the first row to verify changes
Car_data.head(1)
```

## **Step 5: Analyze Car Makes**

We'll analyze the different car makes in the dataset and visualize the number of occurrences for each make.


```python
# Count the number of occurrences of each car make
Car_Makecount = Car_data['Make'].value_counts()
print(Car_Makecount)

# Visualize the number of cars by make
plt.figure(figsize=(10,5))
Car_data['Make'].value_counts().plot(kind='bar', color='navy')
plt.title('Number of Cars by Make')
plt.xlabel('Make')
plt.ylabel('Number of Cars')
plt.show()
```

## **Step 6: Filter Data by Origin**

We'll filter the dataset to include only cars from Asia and Europe, and then visualize the number of cars from these regions.


```python
# Filter the data to include only cars from Asia and Europe
filtered_data = Car_data[Car_data['Origin'].isin(['Asia', 'Europe'])]

# Visualize the number of cars by origin
filtered_data['Origin'].value_counts().plot(kind='bar', color='maroon')
plt.title('Number of Cars by Origin')
plt.xlabel('Origin')
plt.ylabel('Number of Cars')
plt.show()
```

## **Step 7: Clean and Convert Price Data**

We'll clean the 'Price' column by removing non-numeric characters and converting it to an integer data type.


```python
# Create a new column by separating the dollar sign from the 'Invoice' column
Car_data['Price'] = Car_data['Invoice'].str.replace('$','')

# Display the data types
Car_data.dtypes

# Remove non-numeric characters from 'Price' column
Car_data['Price'] = Car_data['Price'].str.replace(",","").replace(" ","")

# Convert 'Price' column to integer
Car_data['Price'] = Car_data['Price'].astype(int)

# Check the data type of 'Price' column
print(Car_data['Price'].dtype)
```

```python
# Show data for cars with weight greater than 4000 pounds
Car_data[Car_data['Weight'] > 4000]

# Use the ~ symbol to remove cars with weight greater than 4000 pounds
Car_data = Car_data[~(Car_data['Weight'] > 4000)]

# Check the shape of the dataset after filtering
Car_data.shape
```

## **Step 9: Adjust MPG City Values**

We'll increase the 'MPG_City' values by 3 for all cars.


```python
# Increase all values in the 'MPG_City' column by 3
Car_data['MPG_City'] = Car_data['MPG_City'].apply(lambda x: x + 3)

# Display the updated dataset
Car_data
```

## **Step 10: Visualize Average Car Prices**

We'll visualize the average prices of cars from Asia and Europe and compare them using a line plot.


```python
# Create a figure with larger size
plt.figure(figsize=(12,8))

# Calculate the average price for cars from Asia and Europe
avg_price = Car_data[Car_data['Origin'].isin(['Asia', 'Europe'])]['Price'].mean()

# Create a line plot for cars from Asia and Europe
sns.lineplot(x='Make', y='Price', data=Car_data[Car_data['Origin'].isin(['Asia', 'Europe'])], color='skyblue', marker='o', hue='Origin')

# Add a horizontal line to indicate the average price
plt.axhline(y=avg_price, color='black', linestyle='dashed')

# Set labels and title
plt.title('Average Car Prices by Make (Asia and Europe)', fontsize=20)
plt.xlabel('Make', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.legend(loc='upper left', fontsize=14)
plt.show()
```

## **Step 11: Analyze Price Distribution**

We'll sort the dataset by price in descending order, select the top 15 car types, and create a bar plot to visualize the price distribution by make and origin.


```python
# Sort the dataframe by Price in descending order
Car_data = Car_data.sort_values(by='Price', ascending=False)

# Select the top 15 car types
Car_data = Car_data.head(15)

# Create a bar plot
sns.barplot(x='Make', y='Price', data=Car_data, hue='Origin', palette='Blues')

# Set labels and titles
plt.title("Price Distribution of Top 15 Car Types by Make and Origin", fontsize=20)
plt.xlabel("Make", fontsize=14)
plt.ylabel("Price", fontsize=14)
plt.xticks(rotation=90)
plt.legend(loc='upper left', fontsize=14)
plt.show()
```

## **Step 12: Analyze Average Cylinder Size**

We'll group the data by car type, calculate the average cylinder size, and visualize it using a bar plot.


```python
# Group the data by car type and calculate the average cylinder size
# Selecting only the 'Cylinders' column ensures we are working with numeric data.
car_data_grouped = Car_data.groupby('Make')[['Cylinders']].mean()

# Sort the data by average cylinder size in descending order
car_data_grouped.sort_values(by='Cylinders', ascending=False, inplace=True)

# Select the top 10 car types by average cylinder size
top_10_car_types = car_data_grouped.head(10)

# Plot size
plt.figure(figsize=(12,8))

# Create a bar plot of the average cylinder size by top 10 car types
sns.barplot(x=top_10_car_types.index, y='Cylinders', palette='PuBu', data=top_10_car_types)

# Set the title and labels
plt.title("Average Cylinder Size by Top 10 Car Types", fontsize=16)
plt.xlabel("Car Type")
plt.ylabel("Average Cylinder Size")
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.show()
```

## **Step 13: Engine Size vs Horsepower**

We'll create scatter plots and regression plots to examine the relationship between engine size and horsepower, categorized by car origin.


```python
# Plot size
plt.figure(figsize=(14,8))

# Create a scatter plot of engine size vs horsepower
sns.scatterplot(x='EngineSize', y='Horsepower', data=Car_data, hue='Origin', palette='Purples')

# Set the title, labels, and size of the title
plt.title("Relation between Engine Size and Horsepower by Origin", fontsize=15)
plt.xlabel("Engine Size", fontsize=12)
plt.ylabel("Horsepower", fontsize=12)
plt.legend(loc='upper left')  # Set location of legend
plt.show()

# Plot size
plt.figure(figsize=(20,10))

# Create a regression plot of engine size vs horsepower
sns.lmplot(x='EngineSize', y='Horsepower', data=Car_data, hue='Origin', palette='YlOrBr', fit_reg=True)

# Set the title and labels
plt.title("Relation between Engine Size and Horsepower by Origin", fontsize=16)
plt.xlabel("Engine Size")
plt.ylabel("Horsepower")
plt.show()
```
